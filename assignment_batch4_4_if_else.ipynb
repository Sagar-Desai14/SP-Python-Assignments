### 40. Check if a number is Armstrong (3-digit).

```python
num=15

# check if number is Armstrong number
sum_of_cubes=0
for digit in str(num):
    sum_of_cubes += int(digit) ** 3
if sum_of_cubes == num:
    print(f"{num} is an Armstrong number")
else:
    print(f"{num} is not an Armstrong number")  
```

### 42. Check if a number is perfect (sum of factors equals number).

```python
num=6
# check if number is perfect(sum of factors equal to number)
sum_of_factors=0
for i in range(1, num):
    if num % i == 0:
        sum_of_factors += i
if sum_of_factors == num:
    print(f"{num} is a Perfect number")
else:
    print(f"{num} is not a Perfect number")
```
