**Q22.** Explain (in a comment) why `convert_temperature()` should be an instance method if it depends on an object’s `scale` (C or F). Optionally, show a small example.

```python
# If convert_temperature() depends on an object's scale (C or F),
# then it MUST be an instance method because:
# - Each object can have a different scale (Celsius or Fahrenheit)
# - The method should use the object's own scale value
# - Static methods do not have access to instance variables like self.scale

class Temperature:
    def __init__(self, value, scale):
        self.value = value      # temperature value
        self.scale = scale      # "C" or "F"

    def convert_temperature(self):
        # Convert based on the object's own scale
        if self.scale == "C":
            return (self.value * 9/5) + 32   # C → F
        elif self.scale == "F":
            return (self.value - 32) * 5/9   # F → C


# Example
t1 = Temperature(30, "C")
t2 = Temperature(86, "F")

print(t1.convert_temperature())   # 30°C → 86°F
print(t2.convert_temperature())   # 86°F → 30°C
```

**Q24.** Make a class `Date` with class method `from_string(cls, s)` to parse `"DD-MM-YYYY"` and return an object.

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    @classmethod
    def from_string(cls, s):
        # split "DD-MM-YYYY"
        d, m, y = s.split("-")
        return cls(int(d), int(m), int(y))  # return Date object

    def show(self):
        print(f"{self.day}-{self.month}-{self.year}")


# Example to run
date1 = Date.from_string("15-02-2025")
date1.show()
```

**Q25.** Create a class `AppConfig` with class variable `DEFAULT_TIMEOUT` and class method to change it. Show that new instances use updated default.

```python
class AppConfig:
    DEFAULT_TIMEOUT = 30   # class variable

    def __init__(self):
        self.timeout = AppConfig.DEFAULT_TIMEOUT

    @classmethod
    def set_timeout(cls, new_timeout):
        cls.DEFAULT_TIMEOUT = new_timeout


# Before changing timeout
a1 = AppConfig()
print("a1 timeout:", a1.timeout)

# Change default timeout using class method
AppConfig.set_timeout(60)

# New instance should use updated default
a2 = AppConfig()
print("a2 timeout:", a2.timeout)
```

**Q26.** Show difference between class variable and instance variable with a simple example in one class.

```python
class Demo:
    class_var = 100   # class variable (shared by all objects)

    def __init__(self, value):
        self.instance_var = value   # instance variable (different for each object)


# Create two objects
d1 = Demo(10)
d2 = Demo(20)

print("Class Variable (d1):", d1.class_var)
print("Class Variable (d2):", d2.class_var)

print("Instance Variable (d1):", d1.instance_var)
print("Instance Variable (d2):", d2.instance_var)

# Changing class variable affects all objects
Demo.class_var = 200

print("After changing class_var:")
print("d1 class_var:", d1.class_var)
print("d2 class_var:", d2.class_var)
```

## G. Decorators (27–30)


**Q27.** Write a simple decorator that prints `"Function started"` before running the function. Test it on a small function.

```python
def my_decorator(func):
    def wrapper():
        print("Function started")
        return func()
    return wrapper


@my_decorator
def say_hello():
    print("Hello!")


# Test
say_hello()
```

**Q28.** Create a decorator that prints `"Function ended"` after running the function. Test it.

```python
def end_decorator(func):
    def wrapper():
        result = func()
        print("Function ended")
        return result
    return wrapper


@end_decorator
def greet():
    print("Welcome!")


# Test
greet()
```

**Q29.** Make a decorator that prints the **name** of the function being called (use `func.__name__`).

```python
def name_decorator(func):
    def wrapper():
        print("Calling function:", func.__name__)
        return func()
    return wrapper


@name_decorator
def hello():
    print("Hello from function!")


# Test
hello()
```

**Q30.** Combine two decorators: one prints `"Start"`, another prints `"End"` around a function. Check the order of messages.

```python
def start_decorator(func):
    def wrapper():
        print("Start")
        return func()
    return wrapper


def end_decorator(func):
    def wrapper():
        result = func()
        print("End")
        return result
    return wrapper


@start_decorator
@end_decorator
def my_function():
    print("Running function...")


# Test
my_function()
```
