## B. Calling One Method Inside Another (6–10)


**Q6.** Create a class `Demo` with two methods: `method1()` prints `"Method 1"`, `method2()` calls `method1()` and then prints `"Method 2"`.

```python
class Demo:
    def method1(self):
        print("I am method 1")

    def method2(self):
        self.method1()
        print("I am method 2")
        

obj= Demo()
obj.method2()
```

**Q7.** Make a class `Math` with `square(n)` and `double_square(n)` that calls `square(n)` and doubles the result.

```python
class Math:
    def square(self,num):
        return num**2

    def double_square(self,num):
        result = self.square(num)
        print(result * 2)

obj = Math()
obj.double_square(4)
```

**Q8.** Create a class `Car` with `start_engine()` and `drive()` where `drive()` calls `start_engine()` first, then prints `"Driving..."`.

```python
class Car:
    def start_engine(self):
        print("Engine Started >>>>>>")

    def drive(self):
        self.start_engine()      # call engine first
        print("Driving......... (-^-)")
        
obj = Car()
obj.drive()
```

**Q9.** Write a class `Student` with `get_name()` and `show_details()` that calls `get_name()` inside it and prints the name.

```python
class Student:
    def get_name(self, name):
        return name

    def show_details(self, name):
        student_name = self.get_name(name)   # calling get_name()
        print("Student Name:", student_name)

obj = Student()
obj.show_details("Sagar")
```

**Q10.** Create a class `Calculator` with `add(a,b)` and `sum_and_print(a,b)` that calls `add()` and prints the result.

```python
class Calculator:
    def add(self,a,b):
        return a+b
    
    
    def sum_and_print(self,a,b):
        total = self.add(a,b)
        print(total)

obj = Calculator()
obj.sum_and_print(2,3)
```

**Q12.** Make a class `Book` with `__init__(self, title, author)` and a method `show_info()` to display both.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def show_info(self):
        print("Title:", self.title)
        print("Author:", self.author)

# creating object
obj = Book("Python Basics", "John Doe")
obj.show_info()
```
