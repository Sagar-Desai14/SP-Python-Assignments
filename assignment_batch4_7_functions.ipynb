### What are keyword arguments? Give an example.

Keyword arguments are arguments where you specify the parameter name while calling the function.
Because of this, the order does not matter.

```python
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=20, name="Sagar")
```

## Explanation:

You wrote age=20 and name="Sagar"

So Python knows exactly which value goes to which parameter

Even though you changed the order, it still works correctly

### What does *args mean in Python? Write a function that accepts any number of values using *args and prints them.

*args allows a function to accept any number of positional arguments.


It collects all extra values into a tuple.

```python
def show_values(*args):
    for value in args:
        print(value)

show_values(10, 20, 30, "Hello", 5.5)
```

### What does **kwargs mean? Write a function that prints all keyword arguments using **kwargs.

**kwargs allows a function to accept any number of keyword arguments (i.e., arguments with names).


It collects all these arguments into a dictionary.

```python
def show_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

show_info(name="Sagar", age=25, city="Mumbai")
```

### Write a function that returns two values: the sum and product of two numbers. Can a Python function return multiple values? How?

Yes! A Python function can return multiple values.


Actually, Python returns them as a tuple, which you can unpack if needed.

```python
def sum_and_product(a, b):
    sum_val = a + b
    product_val = a * b
    return sum_val, product_val  # returns a tuple

# Using the function
s, p = sum_and_product(4, 5)  # tuple unpacking
print("Sum:", s)
print("Product:", p)
```

### Write a function even_or_odd(n) that returns "Even" or "Odd".
